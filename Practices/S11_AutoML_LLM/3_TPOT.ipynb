# Import Libraries

```python
import pandas as pd
import numpy as np

from __future__ import print_function
import sys,tempfile, urllib, os
```

```python
from tpot import TPOTClassifier
```

# Load Data

```python
train_df = pd.read_csv('https://raw.githubusercontent.com/Existanze54/sirius-neural-networks-2024/main/Datasets/patient-survival-prediction/train_preprocessed.csv')
print(train_df.shape)
train_df.head()
```

```python
test_df = pd.read_csv('https://raw.githubusercontent.com/Existanze54/sirius-neural-networks-2024/main/Datasets/patient-survival-prediction/test_preprocessed.csv')
print(test_df.shape)
test_df.head()
```

```python
X_train = train_df.drop("hospital_death", axis=1)
y_train = train_df['hospital_death']
```

```python
X_test = test_df.drop("hospital_death", axis=1)
y_test = test_df['hospital_death']
```

# TPOT Model building

Use TPOT classifier to find the best model algorithm that fits the data. TPOT uses genetic algorithm based search internally

We built 10 different model with 10 population size out of which we will select the best model

```python
tpot = TPOTClassifier(generations=10, population_size=10, verbosity=3)
```

```python
# GPU is mandatory (?)

# TPOT (Tree-based Pipeline Optimization Tool) не имеет встроенной поддержки GPU.
#  TPOT использует библиотеку DEAP для реализации генетических алгоритмов,
#  а сама оптимизация моделей в TPOT основана на Scikit-learn, который также не поддерживает GPU.
```

```python
%%time
tpot.fit(X_train, y_train)
```

```python
print(tpot.score(X_test, y_test)) # By default, accuracy is used for classification and mean squared error (MSE) is used for regression.
```

# Checking scores after cross validation

```python
import multiprocessing

if __name__ == '__main__':
    multiprocessing.set_start_method('forkserver', force=True)
    tpot =  TPOTClassifier(generations=10, population_size=20, verbosity=2, n_jobs=20, random_state=123)
    tpot.fit(X_train, y_train)
```

```python
print(tpot.score(X_test, y_test))
```

Export the scikit learn code for best classifier and dispay the code

```python
tpot.export('tpot_churn_pipeline.py')
```
