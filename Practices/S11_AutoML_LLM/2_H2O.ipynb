# Import Libraries & Initialize the library

```python
import h2o
```

```python
h2o.init()
```

```python
from h2o.automl import H2OAutoML
```

# Load Data

```python
train_data = h2o.import_file('https://raw.githubusercontent.com/Existanze54/sirius-neural-networks-2024/main/Datasets/patient-survival-prediction/train_preprocessed.csv')
```

```python
test_data = h2o.import_file('https://raw.githubusercontent.com/Existanze54/sirius-neural-networks-2024/main/Datasets/patient-survival-prediction/test_preprocessed.csv')
```

```python
train_data.head()
```

```python
train_data.types
```

```python
train_data.describe()
```

```python
# We have detected that your response column has only 2 unique values (0/1). If you wish to train a binary model instead of a regression model, convert your target column to categorical before training.

# Set response column as a factor
train_data["hospital_death"] = train_data["hospital_death"].asfactor()
test_data["hospital_death"] = test_data["hospital_death"].asfactor()
```

```python
y = "hospital_death"
x = train_data.columns
x.remove(y)
```

# Create model object

maximum model= 10

nfold=0 means no cross validation (default=5)

exclude_algorithms = stacking and dL both allowed

for entire seen output use verbosity ='debug'

```python
aml = H2OAutoML(max_models = 10, seed = 10, exclude_algos = ["StackedEnsemble", "DeepLearning"], verbosity="info", nfolds=0)
```

```python
!nvidia-smi
```

```python
%%time
aml.train(x = x, y = y, training_frame = train_data)
```

```python
leaderboard = aml.leaderboard
leaderboard.head()
```

# Prediction on test data

```python
data_pred = aml.leader.predict(test_data)
```

```python
data_pred.head()
## Probabilty of 0 and 1 for head(10) records
```

# Check performance metrics

```python
aml.leader.model_performance(test_data)
```

```python
model_ids = list(aml.leaderboard['model_id'].as_data_frame().iloc[:,0])
model_ids
```

# Check Parameters use for xgboost model

```python
final = h2o.get_model([i for i in model_ids if "XGBoost" in i][0])
final
```

```python
final.params
```
